```python
days=list(range(1,31,3))
print(days)
```

```python
temps=[23,43,2,33,62,55,44,33,22,11]
```

```python
plt.plot(days,temps)
plt.show()
```

```python
plt.plot(days,temps,'xg')
```

```python
temp_min=[22,35,24,55,67,43,11,2,34,15]
temp_max=[27,36,26,57,68,44,19,7,37,16]
```

```python
plt.plot(days,temp_min,'or',days,temp_max,'xr')
plt.show()
```

```python
plt.plot(days,temp_min,days,temp_max)
plt.show()
```

```python
plt.plot(days,temp_min,days,temp_max,days,temp_min,'ob',days,temp_max,'xr')
plt.xlabel('Days')
plt.ylabel('Temperatur')
plt.show()
```

## plt.plot(____sequence1___,___sequence2____,____for points___,.........)

Sequence1: it consists of various elements and it treat as x-axis.
Sequence2: it consists of various elements and it treat as y-axis.
for points:for pointing or specifying certain points on the graph we use special symbols like o and x and associte these symbols with the specific color like for red colur we specify r i.e. 'or'.

## Two lines on same graph :

Two lines on same graph can be plotted easily just by repeating the same sequence specified in general form.

## Spines:

                Spines are the line connecting the axis tick marks and noting the boundaries of data area. They can be placed at arbitrary positions.
                The default position is('outward',o)
## Ticks:
                Ticks are the small small tiny lines on the x axis and y axis in above figure.

## plt.gca()

It gives the instance of the axis.
All the instances of current plot is given by the gca().

```python
plt.plot(days,temp_min,days,temp_max,days,temp_min,'or',days,temp_max,'xr')
ax=plt.gca()
ax.spines['right'].set_visible(False)
plt.show()
```

```python
plt.plot(days,temp_min,days,temp_max,days,temp_min,'or',days,temp_max,'xr')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```

```python
plt.plot(days,temp_min,days,temp_max,days,temp_min,'or',days,temp_max,'xr')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_bounds(20,70)
plt.show()
```

## Ticks :

```python
plt.plot(days,temp_min,days,temp_max,days,temp_min,'or',days,temp_max,'xr')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xticks([0,1])
plt.show()
```

```python
plt.plot(days,temp_min,days,temp_max,days,temp_min,'or',days,temp_max,'xr')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xticks(list(range(0,40,7)),['W1','W2','W3','W4','W5'])
plt.show()
```
