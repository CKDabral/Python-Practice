### 3.) np.empty_like(array_like,dtype=None,order='K',subok=True)

#Return a new array with same shape and type as a given array.
array_like:a:The shape and data-type of a define these same attrbute of returne array
dtype:optional parameter and Override the data type of result.
oder:{'C','F','A',or 'K'} it is also optional.Override the memory layout of result.
    'C' means C-order
    'F' means F-order
    'A' means 'F' if a Fortran contiguous,'C' otherwise
    'K' means match the layout of array_like as closely as possible.
subok:
    bool,optional:
    If True,then newly created array will use the sub-class type of 'a',otherwise it wil be a base-class array.Defaults to True.
    

```python
a=[[1,2,3],[4,5,6]]
b=np.empty_like(a)
print('empty_like array(b) is: {}\nSimple array(a) is {}'.format(b,a))
```

```python
print('Shape of b and type of b is:',b.shape,type(b))
```

```python
print('Shape of a and type of a is:',a.shape,type(a))
```

### 10.) np.full_like(a, fill_value,dtype=None,order='K',subok=True)

Returns a full array with same shape and type s a given array.

Parameters:

a:array_like:The shape and data-type of a define these same attributes of returned array.

fill_value:scalar: Fill value.

dtype:data-type, optional: Overrides the data type of the result.

```python
x=np.arange(6,dtype=int)
np.full_like(x,1)
```

```python
#Now in this we will passing fill_ value of float type but the answer will be of int type this is because we do not explicitly mention the data type. So it inherit the data type of x which is int hence the answer comes in int type.
np.full_like(x,3.4)
```
